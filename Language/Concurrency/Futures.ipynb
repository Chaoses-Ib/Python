```python
import asyncio

async def f() -> str:
    return '1'
fut = asyncio.create_task(f())
```

```python
# /// script
# requires-python = ">=3.8"
# dependencies = [
#     "wrapt~=2.0",
# ]
# ///
import wrapt.proxies

fut = asyncio.Future()
r = wrapt.proxies.LazyObjectProxy(fut.result, interface=str)

try:
    r.removesuffix('')
except asyncio.InvalidStateError as e:
    print(e)

fut.set_result('1')
print(r.removesuffix(''))
```
