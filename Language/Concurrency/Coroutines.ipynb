```python
import types

@types.coroutine
def f():
    print(1)
    if False:
        yield 2
    print(3)
f()
```

```python
import asyncio

async def f():
    print(1)
    await asyncio.sleep(0)
    print(2)
f().send(None)
```

```python
async def f():
    print(1)
    print(2)
    return 3
try:
    f().send(None)
except StopIteration as e:
    print(e.value)
```
